```python
from pathlib import Path 
import pandas as pd 
import numpy as np 

dataset_path = Path().absolute() / "Data"
```

```python
def retrieve_data(path: Path) -> pd.DataFrame :

    return pd.read_csv(path)
```

```python
customers = retrieve_data(dataset_path / "noahs-customers.csv")

noahs_products = retrieve_data(dataset_path / "noahs-products.csv")

noahs_orders = retrieve_data(dataset_path / "noahs-orders.csv")
orders_items = retrieve_data(dataset_path / "noahs-orders_items.csv")
```

```python
noahs_products['category'] = noahs_products['sku'].apply(lambda x: x[:3])
```

```python
noahs_products[noahs_products['category'] == 'HOM']
```

```python
orders_items
```

```python
orders_items = orders_items.merge(noahs_products[['sku', 'wholesale_cost']], how = 'left', on = 'sku')
```

```python
orders_items['line_item_unit'] = orders_items['qty'] * orders_items['unit_price']
orders_items['line_item_wholesale'] = orders_items['qty'] * orders_items['wholesale_cost']
```

```python
orders_items
```

```python
orders_items = orders_items[['orderid', 'line_item_unit', 'line_item_wholesale']]
```

```python
orders_items
```

```python
orders_items = orders_items.groupby('orderid').sum()
```

```python
orders_items.reset_index(inplace = True)
```

```python
orders_items
```

```python
orders_items['profit_or_loss'] = orders_items['line_item_unit'] - orders_items['line_item_wholesale']
```

```python
orders_items.sort_values('profit_or_loss', axis = 0, ascending = True)
```

```python
loss_makers = orders_items[orders_items['profit_or_loss'] < 0]
```

```python
loss_makers.reset_index(drop = True, inplace = True)
```

```python
loss_makers
```

```python
noahs_loss_orders = noahs_orders[noahs_orders['orderid'].isin(loss_makers['orderid'])]
```

```python
loss_makers = loss_makers.merge(noahs_loss_orders[['orderid', 'customerid']], how = 'left', on = 'orderid')
```

```python
loss_makers.head(5)
```

```python
loss_makers = loss_makers.merge(customers[['customerid', 'name', 'phone']], how = 'left', on = 'customerid')
```

```python
loss_makers
```

```python
couponers = loss_makers[['name', 'phone', 'profit_or_loss']]
```

```python
couponers
```

```python
couponers = couponers.groupby(['name', 'phone']).sum()
couponers.reset_index(inplace = True)
```

```python
couponers.sort_values('profit_or_loss', ascending= True, axis = 0)
```
