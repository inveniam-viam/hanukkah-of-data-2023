```python
from pathlib import Path 
import pandas as pd 
import numpy as np 

dataset_path = Path().absolute() / "Data"
```

```python
def retrieve_data(path: Path) -> pd.DataFrame :

    return pd.read_csv(path)
```

```python
customers = retrieve_data(dataset_path / "noahs-customers.csv")

noahs_products = retrieve_data(dataset_path / "noahs-products.csv")

noahs_orders = retrieve_data(dataset_path / "noahs-orders.csv")
orders_items = retrieve_data(dataset_path / "noahs-orders_items.csv")
```

```python
collectibles = noahs_products[noahs_products['sku'].str.startswith('COL')]
```

```python
collectibles
```

```python
collectibles.reset_index(drop = True, inplace = True)
```

```python
collectibles
```

```python
orders_items = orders_items.merge(noahs_products[['sku', 'desc']], how = 'left', on = 'sku')
```

```python
noahs_orders
```

```python
noahs_orders = noahs_orders.merge(orders_items[['orderid', 'sku', 'desc']], how = 'left', on = 'orderid')
```

```python
noahs_orders
```

```python
noahs_col_orders = noahs_orders[noahs_orders['sku'].str.startswith('COL')]
```

```python
noahs_col_orders.reset_index(drop = True, inplace = True)
```

```python
noahs_col_orders
```

```python
col_cust_sku_count = noahs_col_orders[['customerid', 'sku']].groupby('customerid').count()
```

```python
col_cust_sku_count.reset_index(inplace = True)
```

```python
col_cust_sku_count.sort_values('sku', ascending = False, axis = 0)
```
